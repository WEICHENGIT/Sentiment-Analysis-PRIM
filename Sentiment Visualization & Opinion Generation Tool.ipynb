## Demo 1
## Byte level Sentiment Visualization

Here is the API to sentiment visualization tool at byte level. The text will be colored from green to red, when the sentiment varies from positive to negative. 
<img align="left", src="colorbar.png"><br><br><br><br><br><br><br><br>
You can try with your own text to analyse by specifying `any_input` below.


```python
from visualize import visualize

movie_review = "25 August 2003 League of Extraordinary Gentlemen: Sean Connery is one of \
the all time greats and I have been a fan of his since the 1950's. I went \
to this movie because Sean Connery was the main actor. I had not read \
reviews or had any prior knowledge of the movie. The movie surprised me \
quite a bit. The scenery and sights were spectacular, but the plot was \
unreal to the point of being ridiculous. In my mind this was not one of \
his better movies it could be the worst. Why he chose to be in this movie \
is a mystery. For me, going to this movie was a waste of my time. I will \
continue to go to his movies and add his movies to my video collection. \
But I can't see wasting money to put this movie in my collection."

product_review = "You don't have to do anything to get the update; it will be sent automatically by Amazon. \
You may see the indicator light on your Echo will pulse blue as the update is installed. \
The representative I spoke with said you may get it more quickly by not using your Echo for several hours. \
I hit the mute button on mine and received the update within a few hours. \
I'm very impressed by Amazon's quick response to this issue!"

## the text that you'd like to analyse here
any_input = " "

for text in [movie_review, product_review, any_input]:
    visualize(text=text, load_model="e0.pt", model_dir="experiments\mlstm_paper\model", neuron=2388, cuda=False)
```

## Demo 2
## Generating text with customized sentiment polarity

This API is able to automatically generate text just according to the begining of a text snippet you specified.
You can customize the text length by specifying the `seq_length` parameter, and customize the sentiment polarity by setting `overwrite` to +1/-1/None respectively for positive/negative/randomly.

#### With Amazon model

```python
from visualize import visualize

text_initial = "When I received this product yesterday"
visualize(text=text_initial, seq_length=256, overwrite=-1, generate=True,\
          load_model="e0.pt", model_dir="experiments\mlstm_paper\model", neuron=2388, cuda=False)
```

#### With Bloomberg model

```python
from visualize import visualize

text_initial = "French banks are"
visualize(text=text_initial, seq_length=256, overwrite=None, generate=True,\
          load_model="bloomberg.pt", model_dir="experiments\mlstm\model", cuda=False)
```
